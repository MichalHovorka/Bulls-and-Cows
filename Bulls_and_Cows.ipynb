```
# Bulls and Cows project 11 - Michal Hovorka
# Created in colab research google tool

# Definovani promennych
secret_num_size = 4
max_count_guesses = 8
remaining_tips = max_count_guesses

# Vygenerování tajného čísla bez duplicitních znaků a nezačínající nulou
import random
set_digits = set(range(10))
first_digit = random.randint(1, 9)
last_3 = random.sample(set_digits - {first_digit}, 3)
secret_number = (str(first_digit) + ''.join(map(str, last_3)))

# zobrazení vygenerovaného tajného čísla pouze pro potřeby testování, jinak zahashovat(#)
print("!!!!! Only for testing - secret number is: " + str(secret_number) + " !!!!!")

# Uvitani
print(f'''

Hi there!

-----------------------------------------------

I've generated a secret random number for you
It contains {secret_num_size} unique set_digits from 0-9
Let's play a bulls and cows game
-----------------------------------------------
''')

# Zacatek whilecyclu vcetne validaci a hlavni logiky hry
while remaining_tips <= max_count_guesses:

    # Vstup od hrace
    player_tip = input("Please enter a number: ")

    # Kontrola zdali bylo zadano cele cislo
    try:
        tmp = int(player_tip)
    except:
        print("You didn't enter an integer, please try again")
        continue

    # Kontrola na delku zadaneho cisla
    if len(str(player_tip)) != secret_num_size:
        print("Your quess must be " + str(secret_num_size) + " " + "digits long")
        continue

    # Hlavni logika hry
    if player_tip == secret_number:
        print("Correct, you've guessed the right number !")
        break
    else:
        bulls = 0
        cows = 0
        for b in range(secret_num_size):
            if player_tip[b] == secret_number[b]:
                bulls += 1
        for c in range(secret_num_size):
            if player_tip[c] in secret_number and player_tip[c] != secret_number[c]:
                cows += 1
        # Vracene ohodnoceni, ktere bere ohled na jednotne a mnozne cislo ve vystupu
        if  bulls <= 1:
            print("Bull: " + str(bulls))
        else:
            print("Bulls: " + str(bulls))

        if  cows <= 1:
            print("Cow: " +str(cows))
        else:
            print("Cows: " +str(cows))

        # Odpocet zbyvajicich kol hry
        if remaining_tips > 1:
            print("Remaining tips: " + str(remaining_tips - 1))
            remaining_tips -= 1
        else:
            print("You lost !")
            break
```
